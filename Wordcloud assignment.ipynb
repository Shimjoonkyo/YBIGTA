```python
from wordcloud import WordCloud, STOPWORDS as stopwords
```

```python
stopwords
```

```python
mobydick = open("C:\work\moby-dick.txt","r").read()
mobydick
```

```python
mobydick_wc = WordCloud(background_color = "white",
                  collocations = False, 
                  stopwords = stopwords)

mobydick_wc.generate(mobydick)
```

```python
mobydick_wc.words_
```

```python
mobydick_wc.to_image()
```
